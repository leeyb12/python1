```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud

#   
text = """
The ruling came about two months after acting president Choi on Jan. 1 appointed two out of three candidates nominated by the National Assembly as the Constitutional Court justices.
As a result, Chung Kye-sun, then head of the Seoul Western District Court; and attorney Cho Han-chang filled in the judicial vacancies of the Constitutional Court in January. Ma, however, has yet to be appointed.
Ma, a left-leaning justice, is a senior judge of the Seoul Western District Court.
National Assembly Speaker Rep. Woo Won-shik filed the case with the Constitutional Court in January. The hearings on competence disputes ended on Feb. 10.
"""

# 
wordcloud = WordCloud(width=800, height=400, background_color='black').generate(text)

# 
plt.figure(figsize=(10,5))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```

```python
#   
import matplotlib.pyplot as plt #    
from wordcloud import WordCloud #    
import matplotlib.font_manager as fm
from konlpy.tag import Komoran #  

#   
font_path = "./ref/D2Coding-Ver1.3.2-20180524.ttf"

#    
text = """
         ,     .
               .
                .
          .
                   .
               .
               .
          .        .
           .
               .
                .
          .
                   .
               .
               .
          .        .
           .
     .
           1  (FPV) .
         ,      .
       ,       .
              .
              .
""" # wordcloud    (   )

# data = [t for t in text if bool(t.strip())]
# print(data)
komoran = Komoran()
nouns = komoran.nouns(text)
nouns_text = " ".join(nouns)
print(nouns_text)

#   
# WordCloud    :
# width=800:    800  
# height=400 :    400  
# background_color='white':     
wordcloud = WordCloud(font_path, width=800, height=400, background_color='white').generate(text)

# 
plt.figure(figsize=(10,5)) #     ( 10,  5)
plt.imshow(wordcloud, interpolation="bilinear") #     
plt.axis("off") # x, y    
plt.savefig("word2.png")
plt.show() #   
```

```python
from collections import Counter
word_counts = Counter(nouns_text.split())
print(word_counts)
```
