- anagrams: 단어 조합과 암호화와 관련된 데이터셋으로 보이며, 텍스트 데이터 처리나 패턴 분석에 유용합니다.
  
- anscombe: Anscombe's Quartet를 포함하며, 데이터 시각화의 중요성을 보여주는 통계 데이터입니다.

- attention: 주의력에 따른 변화나 과정을 나타내는 실험 데이터일 가능성이 있습니다.

- brain_networks: 뇌 연결 네트워크를 나타내는 데이터로, 신경과학적 시각화를 연습할 때 사용됩니다.

- car_crashes: 자동차 사고와 관련된 통계 데이터를 포함하며, 위험도 분석 및 상관관계를 시각화하는 데 활용됩니다.

- diamonds: 다이아몬드의 특성과 가격 데이터를 다루며, 회귀분석 및 분포 시각화에 적합합니다.

- dots: 점과 관련된 데이터로, 시계열이나 애니메이션 시각화를 연습할 수 있습니다.

- dowjones: 다우존스 산업평균지수의 경제 데이터로, 금융 및 경제 시각화에 사용됩니다.

- exercise: 운동 습관과 효과를 분석하는 데이터로 보이며, 건강 관련 시각화에 유용합니다.

- flights: 항공편 승객 수 데이터를 다루며, 시계열 시각화를 배우는 데 적합합니다.

- -fmri: 기능적 자기공명영상 데이터를 포함하며, 신경과학적 분석 및 시각화에 사용됩니다.

- geyser: 간헐천의 분출 간격과 지속 시간을 나타내는 데이터로, 분포 시각화에 유용합니다.

- glue: 언어 데이터 분석과 관련된 데이터셋일 가능성이 큽니다.

- healthexp: 건강 지출 데이터를 나타내며, 경제 및 건강 데이터를 시각화할 때 활용됩니다.

- iris: 붓꽃 데이터로 유명하며, 머신러닝 및 분류 문제를 연습할 때 자주 사용됩니다.

- mpg: 자동차의 연료 효율성을 나타내는 데이터로, 상관관계 분석에 적합합니다.

- penguins: 펭귄의 생태학적 특성을 다룬 데이터로, 분류 및 군집화 연습에 유용합니다.

- planets: 외계 행성 데이터를 포함하며, 천문학 데이터 시각화를 연습할 때 사용됩니다.

- seaice: 해빙 면적 데이터를 나타내며, 환경 및 기후 변화 시각화에 유용합니다.

- taxis: 뉴욕시 택시 데이터를 다루며, 도시 교통 데이터를 분석할 때 사용됩니다.

- tips: 팁 문화 및 레스토랑 데이터를 다루며, 분포 및 관계 분석에 적합합니다.

- titanic: 타이타닉 생존자 데이터를 포함하며, 생존 분석 및 머신러닝 연습에 유용합니다.

```python
import seaborn as sns
import matplotlib.pyplot as plt
tips = sns.load_dataset('tips')

fig = plt.figure(figsize=(15,5))
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)

sns.stripplot(x='day', y='tip', hue='sex', data=tips, alpha = 0.7, ax=ax1)
sns.swarmplot(x='day', y='tip', hue='sex', data=tips, palette='Set2', alpha = 0.7,ax=ax2)
ax1.set_title('Strip plot 으로 표현한 날짜별로 분류한 성별과의 팁 관계')
ax2.set_title('Swarm plot 으로 표현한 날짜별로 분류한 성별과의 팁 관계')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
tips = sns.load_dataset('tips')

fig = plt.figure(figsize=(15,5))
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)

sns.countplot(x='time', data=tips, ax=ax1)
sns.countplot(x='day', hue='day', data=tips, palette='Set2', ax=ax2)
ax1.set_title('Countplot 시간에 따른 팁 빈도')
ax2.set_title('Countplot 시간과 날짜에 따른 빈도차트')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
tips = sns.load_dataset('tips')

fig = plt.figure(figsize=(15,5))
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)

sns.regplot(x="total_bill", y="tip",data=tips, color="blue", scatter_kws={"s":50, "alpha":0.5}, line_kws={"linestyle":"--","color":"red"}, ax=ax1)
sns.regplot(x="total_bill", y="tip",data=tips, color="blue", scatter_kws={"s":50, "alpha":0.5}, fit_reg=False, ax=ax2)

fig.suptitle("Scatter plots 선형회귀의 ON / OFF 비교")
ax1.set_title('선형회귀 ON')
ax2.set_title('선형회귀 OFF')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
iris = sns.load_dataset('iris')

sns.pairplot(iris, hue="species", markers=["o","s","D"] )

plt.show()
```

```python
sns.boxplot(x="species", y="sepal_length", data=iris)
plt.title("꽃받침 길이와 종의 관계")
plt.show()
```

```python
sns.violinplot(x="species", y="petal_length",data=iris, hue="species", palette="muted")
plt.title("종에 따른 꽃잎의 길이 분포")
plt.show()
```

```python
fig = plt.figure(figsize=(12,4))
ax1 = fig.add_subplot(1,3,1)
ax2 = fig.add_subplot(1,3,2)
ax3 = fig.add_subplot(1,3,3)

sns.boxplot(x="species", y="sepal_length", data=iris, ax=ax1)
sns.violinplot(x="species", y="petal_length",data=iris, hue="species", palette="muted", ax=ax2)
sns.histplot(data=iris, x="petal_length", hue="species", kde=True, element="step", palette="Set2", ax=ax3)


ax1.set_title("꽃받침 길이와 종의 관계")
ax2.set_title("종에 따른 꽃잎의 길이 분포1")
ax3.set_title("종에 따른 꽃잎의 길이 분포2")

sns.lmplot(x="sepal_length", y="sepal_width", hue="species", data=iris, markers=["o","s","D"], height=6, aspect=1.2)
plt.title("꽃받침 길이와 너비의 관계")

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
exercise = sns.load_dataset('exercise')
exercise.sample(5)
```

```python
# point plot 
plt.figure(figsize=(7,4))
sns.pointplot(x="time", y="pulse", hue="kind", data=exercise, capsize=0.1)
plt.title("심박수와 운동종류와의 관계")
plt.show()
```

```python
# box plot 
plt.figure(figsize=(7,4))
sns.boxplot(x='time', y='pulse', hue="kind", data=exercise, palette="Set2")
plt. title("심박수와 시간별 운동종류와의 관계")
plt.show()
```

```python
# swarm plot
plt.figure(figsize=(7,4))
sns.swarmplot(x='time', y='pulse', hue="kind", data=exercise, dodge=True, palette="Set1")
plt.title("시간과 운동 유형에 따른 개별 맥박 측정")

plt.show()
```
