```python
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
# import matplotlib.dates as mdates

import pandas as pd
from datetime import datetime
```

```python
# font  D2coding
d2_path = "./ref/D2Coding-Ver1.3.2-20180524.ttf"
fm.fontManager.addfont(d2_path)
matplotlib.rcParams["font.family"] = "D2Coding"
```

```python
file_path ="./ref/__.csv"
ori = pd.read_csv(file_path, encoding="EUC-KR") # utf-8
ori.columns = ["category","date", "internet", "mobile"]
ori.tail()
```

```python
print("")
print(ori['date'].tail())
ori['date'] = ori['date'].map(lambda x: x.strip(' p)'))
print("")
print(ori['date'].tail())
```

```python
ori['date'] = pd.to_datetime(ori["date"], format='%Y.%m')
ori.dtypes
ori.tail()
```

```python
ori[130:170]
```

```python
ex1_data = ori.copy()
# ex1_data[ex1_data['category']== ''] 
obj = ['  ','','','','','','','  ']
ex1_data = ex1_data[ex1_data['category'].isin(obj)]
ex1_data.sample(5)
```

```python
fig , ax = plt.subplots(figsize=(10,5))
category = obj # ['  ','','','','','','','  ']

for cate in category:
    category_data = ex1_data[ex1_data['category'] ==cate]
    x_data = category_data['internet']
    y_data = category_data['mobile']
    ax.scatter(x_data , y_data, s = 20, label = cate)
    
ax.set_xlabel(" ")
ax.set_ylabel(" ")
ax.legend(loc="best")
ax.plot([0,1000000],[0,1000000],'k-',alpha=0.8)
plt.show()
```

```python
fig , ax = plt.subplots(1,2,figsize=(15,5))
category = obj # ['  ','','','','','','','  ']

for cate in category:
    category_data = ex1_data[ex1_data['category'] ==cate]
    x_data = category_data['internet']
    y_data = category_data['mobile']
    ax[0].scatter(x_data , y_data, s = 20, label = cate)
    
ax[0].set_xlabel(" ")
ax[0].set_ylabel(" ")
ax[0].legend(loc="best")
ax[0].plot([0,1000000],[0,1000000],'k-',alpha=0.8)


tdata = category_data['date']
second = ex1_data[ex1_data['category']==obj[0]]
idata = second['internet']
mdata = second['mobile']

ax[1].scatter(tdata, idata, s = 20, label = obj[0]+ '()')
ax[1].scatter(tdata, mdata, s = 20, label = obj[0]+ '()')
ax[1].legend

plt.show()
```

```python
fig, ax = plt.subplots(1,2,figsize=(25,10))
category = obj 

for cate in category:
    category_data = ex1_data[ex1_data['category'] ==cate]
    x_data = category_data['internet']
    y_data = category_data['mobile']
    ax[0].bar(x_data, y_data, width = 100, label = cate)

ax[0].set_xlabel(" ")
ax[0].set_ylabel(" ")
ax[0].legend(loc="best")

tdata = category_data['date']
second = ex1_data[ex1_data['category']==obj[0]]
idata = second['internet']
mdata = second['mobile']

ax[1].bar(tdata, idata, width=20 , label = obj[0]+ '()')
ax[1].bar(tdata, mdata, width=20, label = obj[0]+ '()')
ax[1].legend()

plt.show()
```
