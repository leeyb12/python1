# 내장함수

```python
# abs * : 어떤 숫자를 입력받았을 때 그 숫자의 절댓값을 리턴하는 함수
print(abs(-3))
print(abs(-1.2))
```

```python
# all * : 반복 가능한 데이터 x를 입력값으로 받으며 이 x의 요소가 모두 참이면 True, 거짓이 하나라도 있으면 False를 리턴한다. 
all([1,2,3])
all ([11,22,0,44,55])
all([]) # 특이사항
```

```python
# any * : 반복 가능한 데이터 x를 입력으로 받아 x의 요소 중 하나라도 참이 있으면 True를 리턴하고 x가 모두 거짓일 때만 False를 리턴한다. 즉, all(x)의 반대로 작동한다. 
print(any([1,2,3,0]))
print(any([0,"",[],False]))
print(any([]))
```

```python
# chr : 유니코드 숫자 값을 입력받아 그 코드에 해당하는 문자를 리턴하는 함수
# 아스키코드
print(chr(65))
print(chr(44032))
# 유니코드 - UTF -8(인코딩방식)
print(ord('A'))
print(ord('가'))
```

```python
from konlpy.tag import Okt

# 한글 형태소 분석기
okt = Okt()

positive_words = ["좋다", "행복", "멋지다", "최고", "사랑"]
negative_words = ["싫다", "화나다", "짜증", "별로", "나쁘다"]
text = input("당신의 기분을 한 문장으로 표현해주세요: ")

words = okt.morphs(text, stem=True)
print(words)
```
