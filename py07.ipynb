```python
# all * : 반복 가능한 데이터 x를 입력값으로 받으며 이 x의 요소가 모두 참이면 True, 거짓이 하나라도 있으면 False를 리턴한다. 
all([1,2,3])
all ([11,22,0,44,55])
all([]) # 특이사항
```

```python
# any * : 반복 가능한 데이터 x를 입력으로 받아 x의 요소 중 하나라도 참이 있으면 True를 리턴하고 x가 모두 거짓일 때만 False를 리턴한다. 즉, all(x)의 반대로 작동한다. 
print(any([1,2,3,0]))
print(any([0,"",[],False]))
print(any([]))
```

```python
# filter (함수, 이터러블) *
# positive.py
def positive(l):
    result = []
    for i in l:
        if i > 0:
            result.append(i)
    return result 

print(positive([1, -3, 2, 0, -5, 6]))
```

```python
def pos(l):
    return[i for i in l if i > 0]
print(pos([1,-3,2,0,-5,6]))
```

```python
# isinstance : 첫 번째 인수로 객체, 두 번째 인수로 클래스로 받는다. 
# 입력으로 받은 객체가 그 클래스의 인스턴스인지를 판단하여 참이면 True, 거짓이면 False를 리턴한다. 

# isinstance(object, class) 인스턴스 확인용

class Person: 
    def p(self):
        return("사람입니다.")
class Animal: pass
a = Person()
b = Person()
c = Animal()
print(a.p())
print(b.p())
print(isinstance(a,Person))
print(isinstance(b,Person))
print(isinstance(c,Person))
```

```python
# map (f,iterable): 함수(f)와 반복 가능한 데이터를 입력으로 받는다. 입력받은 데이터의 각 요소에 함수 f를 적용한 결과를 리턴하는 함수

# map **
def two_items(nl):
    result = []
    for n in nl:
        result.append(n*2)
    return result
print(two_items([1,2,3,4]))
```

```python
# 테스트
li = [1,2,3,4,5,6]
def odd(x):
    return x%2 != 0
def even(x):
    return x%2 == 0
def add(x): 
    return x + 10
def mul(x):
    return x * x
def addmul(x):
    return (x+10) * x
print(list(filter(odd,li)))
print(list(filter(even,li)))
print(list(map(add,li)))
print(list(map(mul,li)))
print(list(map(addmul,li)))
```

```python
# 게시판 페이징하기
def get_total_page(num1,num2):
    if num1 % num2 == 0: 
        return num1 // num2 
    else: 
        return num1 // num2 + 1

if __name__ == '__main__':
    while True: 
        try:
            num1 = int(input('게시물의 총 개수를 입력하세요.'))
            num2 = int(input('페이지당 보여줄 게시물 수를 입력하세요.'))

            if num1 < 0 or num2 < 0:
                continue
            else: 
                totalpage = get_total_page(num1,num2)
                print(totalpage)
            break
        except:
            pass
```
